```python
import numpy as np
from fmskill.model import ModelResult
from fmskill.observation import PointObservation
import fmskill.metrics as mtr 
```

```python
fn = '../tests/testdata/TS.dfs0'
mr = ModelResult(fn, 'OresundHD')
print(mr)
print(mr.dfs)
```

```python
fn = '../tests/testdata/smhi_2095_klagshamn.dfs0'
o1 = PointObservation(fn, x=366844.15, y=6154291.6, item=0, name='Klagshamn') 
print(o1)
print(o1.n_points, o1.start_time)
```

```python
mr.add_observation(o1, item='Klagshamn: Surface elevation', validate_eum=False)
mr.observations
```

```python
cc = mr.extract()
cc
```

```python
cc["Klagshamn"].plot_timeseries();
```

```python
cmp = cc["Klagshamn"]
cmp.skill()
```

```python
mtr.rmse(cmp.obs, cmp.mod, unbiased=False)
```
