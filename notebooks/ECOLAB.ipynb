# ECOLAB

This example consists of model results in separate dfs0 files.

```python
from fmskill import ModelResult, PointObservation
```

```python
m_st8 = ModelResult("../tests/testdata/ECOLAB/st8_ECO.dfs0")
m_st8.add_observation(
        PointObservation(
            "../tests/testdata/ECOLAB/obs/st8_waterquality_variables.dfs0", item="IN", variable_name="IN", name="St8_IN"
        ),
        item="St 8 ECO (-1 m): IN, Inorganic nitrogen, g N/m3",
)

m_st8.add_observation(
        PointObservation(
            "../tests/testdata/ECOLAB/obs/st8_waterquality_variables.dfs0", item="IP ", variable_name="IP", name="St8_IP"
        ),
        item="St 8 ECO (-1 m): IP, Inorganic phosphorous, g P/m3",
)

m_st17 = ModelResult("../tests/testdata/ECOLAB/st17_ECO.dfs0")
m_st17.add_observation(
    PointObservation(
        "../tests/testdata/ECOLAB/obs/st17_waterquality_variables_top.dfs0", item="IN",  variable_name="IN", name="St17_IN"
    ),
    item="St 17 ECO (-1 m): IN, Inorganic nitrogen, g N/m3",
)
m_st17.add_observation(
    PointObservation(
        "../tests/testdata/ECOLAB/obs/st17_waterquality_variables_top.dfs0", item="IP ",  variable_name="IP", name="St17_IP"
    ),
    item="St 17 ECO (-1 m): IP, Inorganic phosphorous, g P/m3",
)

cc1 = m_st8.extract()
cc2 = m_st17.extract()
```

```python
cc = cc1 + cc2 # this should also work
```

```python
cc = cc1[0] + cc1[1] + cc2[0] + cc2[1] # this works but is not ideal
```

```python
cc.skill()
```

```python
for c in cc.values():
    c.plot_timeseries()
```
